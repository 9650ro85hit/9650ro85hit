```python
import numpy as np
import pandas as pd
from hmmlearn import hmm
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/prices.csv")
```

```python
data['DATE'] = pd.to_datetime(data['date'])
data.set_index('DATE', inplace=True)
```

```python
observations = data['close'].values.reshape(-1, 1)
```

```python
model = hmm.GaussianHMM(n_components=2, covariance_type="diag", n_iter=1000)
model.fit(observations)
```

```python
hidden_states = model.predict(observations)
```

```python
plt.figure(figsize=(15, 8))

for i in range(model.n_components):
    indices = np.where(hidden_states == i)[0]  # Get indices where hidden state is i
    plt.plot(data.index[indices], observations[indices], ".", label=f"Hidden State {i}")

plt.legend()
plt.title("Hidden Markov Model - Bull and Bear States in Stock Price Time Series")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```
